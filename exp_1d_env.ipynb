```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from environments import RL_ENV, LinearWorldEnv
from agents import RL_Agent, MWUAgentBruteFroce, MWUAgentGrid, QLearningAgent
```

```python
lin_env_small = LinearWorldEnv(grid_size=12, goal_state=9)
lin_env_small.plot()
```

```python
init_state = 3
Qagent = QLearningAgent(lin_env_small, init_state, epsilon=0)
mwu_allagent = MWUAgentBruteFroce(lin_env_small, init_state)
mwu_grid = MWUAgentGrid(lin_env_small, init_state, update_step=3, eta=0.3, gamma=0.92)
```

```python
num_eps = 1000
Qagent_rewards, Qph = Qagent.train(episodes=num_eps)
mwu_allagent_rewards, mwu_allph = mwu_allagent.train(episodes=num_eps)
mwu_grid_rewards, mwu_gridph = mwu_grid.train(episodes=num_eps)
```

```python
lin_env_small.evaluate_policy(policy=Qagent, plot=True)
```

```python
lin_env_small.evaluate_policy(policy=mwu_allagent, plot=True)
```

```python
lin_env_small.evaluate_policy(policy=mwu_grid, plot=True)
```

```python
plt.plot(mwu_allagent_rewards, label='MWU_ALL Agent')
plt.plot(mwu_grid_rewards, label='MWU_Grid Agent')
plt.plot(Qagent_rewards, label='Q-Learning Agent')
plt.xlabel('Episode')
plt.ylabel('Cumulative Reward')
plt.legend()
plt.title('MWU_ALL Agent vs Q-Learning Agent in Gridworld')
plt.show()
```
