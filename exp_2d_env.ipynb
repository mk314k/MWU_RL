```python
from environments import GridWorldEnv
from matplotlib import pyplot as plt
from environments import RL_ENV
import numpy as np
```

```python
grid_env_simple = GridWorldEnv(grid_size=(5, 6), goal_state=(4, 4))
grid_env_simple.plot()
```

```python
from agents import QLearningAgent, MWUAgentGrid, ExpectedSARSAgent, SARSAgent, DynaQAgent, QLearningExpertAgent, MWUAgentQGrid
```

```python
qagent = QLearningAgent(grid_env_simple)
magent = MWUAgentGrid(grid_env_simple, update_step=3)
mqagent = MWUAgentQGrid(grid_env_simple, update_step=3)
qeagent = QLearningExpertAgent(grid_env_simple)
esagent = ExpectedSARSAgent(grid_env_simple)
sagent = SARSAgent(grid_env_simple)
dagent = DynaQAgent(grid_env_simple)
```

```python
agents = [qagent, magent, mqagent, qeagent, esagent, sagent, dagent]
agents_name = ['qagent', 'magent', 'mqagent', 'qeagent', 'esagent', 'sagent', 'dagent']
```

```python
agent_rewards = []
for agent in agents:
    _, _, agent_reward = agent.train(episodes=200, evaluate_policy=True)
    agent_rewards.append(agent_reward)
```

```python
select_agent = {
    'qagent', 
    # 'magent', 
    'mqagent',
    'qeagent',  
    # 'esagent', 
    # 'sagent', 
    # 'dagent'
}
```

```python
for i, aname in enumerate(agents_name):
    if aname in select_agent:
        plt.plot(agent_rewards[i][10:], label=aname)
plt.xlabel('Episode')
plt.ylabel('Cumulative Reward')
plt.legend()
plt.title('Agents in Gridworld')
plt.show()
```
